```python
import requests
from bs4 import BeautifulSoup
from requests import api
from requests.api import head
import csv
import pandas as pd
import numpy as np


url = 'https://www.crunchbase.com/organization/y-combinator/school_financials'
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0", "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "Accept-Language": "en-US,en;q=0.5", "Accept-Encoding": "gzip, deflate", "DNT": "1", "Connection": "close", "Upgrade-Insecure-Requests": "1"}

response = requests.get(url, headers=headers)
html_content = response.content
soup = BeautifulSoup(html_content, 'html.parser')



# <--------------------------------------** Invesment **------------------------------------------------------>



soup5 = soup.find('table', class_='card-grid')
soup6 = soup.find('div', class_="identifier-image-and-label ng-star-inserted")
soup7 = soup.find('a', class_="component--field-formatter field-type-identifier link-accent ng-star-inserted")


Annu_lst = []
for i in soup5:
    annou = i.find_all_next('span', class_='component--field-formatter field-type-date ng-star-inserted')
    for k in annou:
        p4 = k.text
        Annu_lst.append(p4)
orgnization_list1 = Annu_lst[5:26]
orgnization_list1.pop(10)
orgnization_list1

df_date = pd.DataFrame(orgnization_list1,columns = ["Announcement date"])
```

```python
df_date
```

```python
org_lst = []
for i in soup7:
    org_name = i.find_all_next('div', class_='identifier-label')
    for k in org_name:
        p3 =k.text
        org_lst.append(p3)

orgnization_name_funding = pd.DataFrame([org_lst[i:i+2] for i in range(4,len(org_lst),2)],
             columns = ['Orgnization name','Funding round'])
```

```python
orgnization_name_funding
```

```python
lead_lst = []
for j in soup5:
    lead_invest = j.find_all_next('span', class_="component--field-formatter field-type-boolean ng-star-inserted")
    for i in lead_invest:
        p5 = i.text
        lead_lst.append(p5)
df_lead = pd.DataFrame(lead_lst, columns =['Lead Investor'])
df_lead
```

```python
money_lst = []
for k in soup5:
    money_raise = k.find_all_next('span', class_="component--field-formatter field-type-money ng-star-inserted")
    for j in money_raise:
        p7 = j.text
        money_lst.append(p7)
funding_list1 = money_lst[5:26]

funding_list1.pop(4)
funding_list1
    
df_money = pd.DataFrame(funding_list1,columns =['Funding Round'])
df_money
```

```python
financial_table_3_4 = [df_date, orgnization_name_funding, df_lead, df_money]
# frames = df_date, orgnization_name_funding
# df4 = pd.concat(frames)

financial_3_4 = pd.concat(financial_table_3_4, join='outer', axis = 1)
financial_3_4
```
